## **4. Decision Tree Regression**

What is Decision Tree Regression?

Decision Tree Regression is a supervised machine learning algorithm
used for predicting continuous values.

It works by:
- Splitting the dataset into smaller regions
- Predicting the average value of the target variable in each region
- Not assuming any mathematical equation between input and output

--------------------------------------------------

How It Works

1. Choose a feature
2. Find the best split point that minimizes error
3. Split the data into child nodes
4. Repeat the process recursively

Stopping conditions:
- Maximum depth is reached
- Minimum samples per leaf is reached

--------------------------------------------------

Split Criterion (Regression)

For regression problems, the decision tree uses
Mean Squared Error (MSE) to decide the best split.

MSE Formula:

MSE = (1 / n) * Σ (y - y_mean)^2

Where:
- y       : actual value
- y_mean  : predicted (mean) value
- n       : number of samples

The tree selects the split that gives the
maximum reduction in MSE.

--------------------------------------------------

Advantages

- Captures non-linear relationships
- No feature scaling required
- Easy to unders

##**5. Support Vector Regression (SVR)**
What is Support Vector Regression?

Support Vector Regression (SVR) is a supervised machine learning algorithm used to predict continuous numerical values.
It is the regression version of Support Vector Machine (SVM).

Unlike linear regression, which tries to minimize prediction error for all data points, SVR focuses on fitting a function that deviates from the actual values by at most a specified margin.

**ε (Epsilon) – Error Margin**

* This zone is called the ε-insensitive tube.
* ε defines a tolerance zone around the regression line

* Errors inside ε are ignored

* Errors outside ε are penalized


📌 If prediction lies inside the tube → no loss

📌 If prediction lies outside the tube → loss is applied

**Advantages of SVR**

1. Works well with high-dimensional data

2. Robust to outliers

3. Handles non-linear regression effectively

4. Good generalization performance

**Limitations of SVR**

1. Computationally expensive for large datasets

2. Choosing kernel and parameters is critical

3. Less interpretable than linear regression
