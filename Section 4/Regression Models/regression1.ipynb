#**Introduction to Regression**

###**What is Regression?**

Regression is a **supervised machine learning technique** used to predict a **continuous numerical value** based on one or more input features.

* The output is a **real number**, not a category
* It learns the **relationship between input variables (features)** and a **target variable**
* Example: Predicting house prices based on area, number of rooms, and location








---

### Real-World Applications of Regression

Regression is widely used in real-life prediction problems:

* 🏠 House price prediction
* 📈 Stock price forecasting
* 🛒 Demand forecasting in retail
* 🚕 Ride fare estimation
* 🌡️ Weather forecasting (temperature, rainfall)
* 💼 Salary prediction based on experience
* ⚡ Energy consumption prediction


Regression helps businesses and systems **make data-driven decisions** using numerical predictions.

## Mathematical Intuition of Regression

### Relationship between Dependent and Independent Variables

In regression, we model the relationship between:

* **Independent variables (X)** → Input features
* **Dependent variable (y)** → Output/target value

The goal is to learn a function **f(X)** such that:

[
y = f(X)
]

**Example:**

* X = house area, number of rooms
* y = house price

The dependent variable **changes based on** the independent variables.

---



### **Linear Relationship Concept**

In **Linear Regression**, we assume a **linear relationship** between input and output.

For a single feature:


y = mx + c


Where:

* **m** → slope (how much y changes when x changes)
* **c** → intercept (value of y when x = 0)

For multiple features:


y = w1x1 + w2x2 + ... + wnxn + b


This means:

* Each feature contributes **linearly** to the final prediction
* The model fits a **straight line (or plane/hyperplane)** to the data

---

### 🔹 Cost Function Intuition (Error Minimization)

The model makes predictions, but predictions may not be perfect.
The **error** is the difference between:

**Error = Actual value−Predicted value**

To measure overall model performance, we use a **cost function**.

**Most common cost function: Mean Squared Error (MSE)**

**MSE=n1​i=1∑n​(yi​−y^​i​)2**

Why squared error?

* Penalizes large errors more
* Always produces a positive value
* Makes optimization mathematically easier

Goal of regression is to find model parameters (**weights and bias**) that **minimize the cost function**.


###**Regression Evaluation Techniques**
---

Used when the output is continuous

Error-Based Metrics:

🔸 **MAE (Mean Absolute Error)**


*   Measures the average absolute difference between actual and predicted values
*   Treats all errors equally
*   Easy to understand and interpret

🔸 **MSE (Mean Squared Error)**

* Calculates the average of squared errors

*   Penalizes large errors more heavily
*   Useful when large mistakes are costly

🔸 **RMSE (Root Mean Squared Error)**

* Square root of MSE

* Expressed in the same units as the target variable

* More sensitive to large errors than MAE

🔸 **R² Score (Coefficient of Determination)**

* Measures how well the model explains the variance in the data

* Value ranges from 0 to 1

* Higher value indicates a better fit

### **1. Simple Linear Regression**
Simple Linear Regression is a supervised learning algorithm used to model the relationship between one independent variable (X) and one dependent variable (y) using a straight line.

It predicts a continuous value based on a single input feature.

**Equation of a Straight Line**

The mathematical form of Simple Linear Regression is: ***y=mx+c***

Where:

x → Independent variable (input)

y → Dependent variable (output)

m → Slope (rate of change of y with respect to x)

c → Intercept (value of y when x = 0)



**Simple Linear Regression works best when:**
*   Relationship between X and y is linear
*   Data has no extreme outliers
*   Errors are independent
*   Variance of errors is constant (homoscedasticity)

# **2. Multiple Linear Regression**
**Why Multiple Features?**

In real-world problems, the output usually depends on more than one input feature.

Example: House Price Prediction
*   Area
*   Number of rooms
*   Location rating
*   Age of house

Using multiple features helps the model:


*   Capture realistic patterns
*   Improve prediction accuracy
*   Reduce oversimplification

**General Equation**

Multiple Linear Regression models the relationship as:


𝑦 = 𝑤1𝑥1 + 𝑤2𝑥2 + ... + 𝑤𝑛𝑥𝑛 + 𝑏

Where:

𝑥1 , 𝑥2..., 𝑥𝑛  → Input features

𝑤1,𝑤2,...,𝑤𝑛  → Coefficients (weights)

𝑏 → Bias (intercept)

𝑦 → Predicted output

Each feature contributes independently and linearly to the prediction.

🔹 Feature Impact and Coefficients

1. Each coefficient represents the impact of one feature

2. A higher absolute value → stronger influence on output

3. Positive coefficient → direct relationship

4. Negative coefficient → inverse relationship

Example:

𝑤1= 5000 → Price increases by ₹5000 per unit increase in feature 𝑥1
	​


🔹 Multicollinearity (Basic Intuition)

Multicollinearity occurs when two or more input features are highly correlated

# **3. Polynomial Regression**

---

## What is Polynomial Regression?

Polynomial Regression is a type of regression that models the relationship between the independent variable **x** and dependent variable **y** as an **nth-degree polynomial**.

It is used when data shows a **non-linear (curved)** relationship.

Even though the curve is non-linear, the model is **linear in parameters**.



---

## Why Polynomial Regression?

Linear regression fits a straight line, which may not capture complex patterns.

Polynomial regression is used when:
- Data has **curvature**
- Linear regression underfits
- Relationship changes at different values of x

Example shapes:
- U-shaped curve
- Inverted U-shape
- S-shaped curve

---


## Degree of Polynomial

| Degree | Model Behavior |
|------|---------------|
| 1 | Linear Regression |
| 2 | Parabolic curve |
| 3 | S-shaped curve |
| High degree | Very flexible but risky |

##  When to Use Polynomial Regression?

✔ Relationship is non-linear  
✔ Dataset is small or medium  
✔ Pattern has curvature  
✔ Interpretability is required  



##  Regularization in Polynomial Regression

Polynomial regression can easily **overfit** when the degree is high, because higher-power terms can produce very large coefficients.

**Regularization** helps control this by adding a **penalty term** to the loss function.

To reduce overfitting:
- **Ridge Regression** → Penalizes large coefficients
- **Lasso Regression** → Feature selection
- **Elastic Net** → Ridge + Lasso

###**Step 1: Import Libraries**




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

###**Step 2: Load Dataset**



```python
import seaborn as sns

# Load real Auto MPG dataset
df = sns.load_dataset("mpg")

df.head()
```

###**Step 3: Transform Features**




```python
df = df[['horsepower', 'mpg']]
df = df.dropna()   # remove missing values

df.head()
```

###**Step 4: Feature Selection**


```python
X = df[['horsepower']].values
y = df['mpg'].values
```

###**Step 5: Polynomial Feature Transformation**


```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

###**Step 6: Train the Model**




```python
model = LinearRegression()
model.fit(X_poly, y)
```

###**Step 7: Make predictions**

```python
y_pred_train = model.predict(X_poly)
```

###**Step 8: Model evaluation metrics**


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


# Mean Absolute Error
mae = mean_absolute_error(y, y_pred_train)

# Mean Squared Error
mse = mean_squared_error(y, y_pred_train)

# Root Mean Squared Error
rmse = np.sqrt(mse)

# R-squared score
r2 = r2_score(y, y_pred_train)

print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R²  :", r2)
```
