# **Project: Tool-Using Research Agent**

Users often need quick information about different topics. Instead of manually searching and reading multiple sources, we can design an AI agent that automatically retrieves and summarizes relevant information.

A Tool-Using Research Agent is an intelligent software system that can perform research tasks by combining reasoning with the ability to use external tools. Instead of relying only on its internal knowledge, the agent can access tools such as search engines, databases, or APIs to gather relevant information. This allows the agent to answer queries more accurately and efficiently. In a single-agent workflow, one autonomous agent is responsible for understanding the user’s request, deciding which tool to use, retrieving information, and generating the final output.

### **Idea**

The agent receives a research question, then uses a tool (Wikipedia search) to get information and return a summary.


### **Working Principle**

The operation of a tool-using research agent generally involves the following steps:

1. **Input Processing**– The user provides a research question or topic.

2. **Reasoning** – The agent analyzes the query and decides the best approach to obtain the information.

3. **Tool Selection and Usage** – The agent uses an external tool (e.g., Wikipedia search) to gather relevant data.

4. **Information Processing** – The agent extracts and organizes the important details from the retrieved data.

5. **Output Generation** – The final research summary or explanation is presented to the user.

**Step 2: Create the Research Agent**

```python
import wikipedia

class ResearchAgent:

    def __init__(self):
        print("Tool-Using Research Agent Initialized")

    # Reasoning step
    def reason(self, topic):
        print("\nUnderstanding the research query...")
        return f"I will search Wikipedia for information about {topic}"

    # Tool usage
    def use_tool(self, topic):
        print("Using Wikipedia tool to fetch information...\n")
        try:
            summary = wikipedia.summary(topic, sentences=3)
            return summary
        except:
            return "Sorry, I couldn't find information on this topic."

    # Complete workflow
    def run(self, topic):
        plan = self.reason(topic)
        result = self.use_tool(topic)

        print("Plan:", plan)
        print("\nResearch Summary:")
        print(result)

        return result
```

* init() – initialize agent

* reason() – understand query

* use_tool() – call Wikipedia

* run() – execute full workflow

**Step 3: Run the Agent**

```python
agent = ResearchAgent()

agent.run("Artificial Intelligence")
```

```python
agent = ResearchAgent()

agent.run("Movies")
```
