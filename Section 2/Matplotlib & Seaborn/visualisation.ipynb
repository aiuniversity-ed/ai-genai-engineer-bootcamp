## What is Data Visualisation?
Data visualization helps in understanding data through graphical representation.
It allows analysts to identify patterns, trends, relationships, and outliers in datasets.  

 **Examples of data visualizations include:**
- Line graphs  
- Bar charts  
- Histograms  
- Scatter plots  
- Heatmaps  

### Why Data Visualization is Needed

- Large datasets are difficult to understand in raw numerical form
- Visual representation helps identify trends and patterns quickly
- Supports better decision making
- Makes communication of insights more effective


## Types of Data Used in Visualization

Data can be broadly classified into:
- **Numerical data** – Continuous and discrete values
- **Categorical data** – Labels or categories
- **Time-series data** – Data recorded over time


### Choosing the Right Plot

The type of visualization depends on:
- Nature of data
- Number of variables
- Purpose of analysis


## Typical Data Visualization Workflow

1. Collect data
2. Clean and preprocess data
3. Perform exploratory data analysis
4. Create visualizations
5. Interpret and communicate insights


## Libraries Used for Visualization

The following Python libraries are commonly used for data visualization:
- **Matplotlib**
- **Seaborn**


```python
# Importing required libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## Data Visualization using Matplotlib

Matplotlib is a basic plotting library in Python that provides full control over
graphs and visualizations. It is commonly used to create line plots, bar plots,
histograms, and scatter plots.

### Features of Matplotlib

- Highly customizable plots
- Supports multiple plot types
- Suitable for simple and complex visualizations
- Works well with NumPy and Pandas



## 1. Line Plot

A **line plot** is used to visualize trends or changes in data over time or continuous intervals.


```python
# Line Plot using Matplotlib
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]

plt.plot(x, y)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Line Plot Example")
plt.show()
```

## 2. Bar Plot

A **bar plot** is used to compare values across different categories.


```python
# Bar Plot using Matplotlib
categories = ['A', 'B', 'C']
values = [20, 35, 30]

plt.bar(categories, values)
plt.title("Bar Plot Example")
plt.show()
```

## 3. Histogram

A **histogram** shows the distribution of numerical data by grouping values into bins.


```python
# Histogram using Matplotlib
import numpy as np
data = np.random.randn(1000)

plt.hist(data, bins=20)
plt.title("Histogram Example")
plt.show()
```

## 4. Scatter Plot

A **scatter plot** is used to observe relationships between two numerical variables.


```python
# Scatter Plot using Matplotlib
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.show()
```

#Statistical Data Visualization using Seaborn
Seaborn is a high-level Python library built on top of Matplotlib.
It is mainly used for statistical data visualization and works
efficiently with Pandas DataFrames.

### Advantages of Seaborn

- Attractive default themes
- Built-in statistical plots
- Easy handling of categorical data
- Supports multivariate analysis


## Seaborn Dataset

Seaborn provides built-in datasets that are useful for visualization practice.


```python
# Loading dataset
df = sns.load_dataset("tips")
df.head()
```

## Count Plot

A count plot displays the frequency of observations in each
categorical bin.


```python
sns.countplot(x="day", data=df)
plt.title("Count Plot Example")
plt.show()
```

## Box Plot

A **box plot** visualizes the distribution of data and helps detect outliers.


```python
# Box Plot using Seaborn
sns.boxplot(x="day", y="total_bill", data=df)
plt.title("Box Plot Example")
plt.show()
```

## Heatmap

A **heatmap** represents data values using color intensity.
It is commonly used to visualize correlation matrices.


```python
# Heatmap using Seaborn
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

## Scatter Plot with Hue

This plot shows the relationship between two variables along with
a third categorical variable using color.


```python
sns.scatterplot(x="total_bill", y="tip", hue="sex", data=df)
plt.show()
```

## Comparison: Matplotlib vs Seaborn

| Feature | Matplotlib | Seaborn |
|-------|-----------|---------|
| Type | Low-level plotting library | High-level statistical visualization library |
| Built on | Core visualization library | Built on top of Matplotlib |
| Ease of Use | Requires more code | Simple and concise syntax |
| Customization | Highly customizable | Limited compared to Matplotlib |
| Default Styling | Basic | Attractive and informative |
| Statistical Plots | Limited support | Excellent support |
| Data Handling | Works with arrays and lists | Works directly with Pandas DataFrames |
| Use Case | Detailed and custom visualizations | Exploratory Data Analysis (EDA) |
| Learning Curve | Steeper | Easier |
