# **Array Attributes in NumPy**

### **Why Array Attributes Matter**



*   Array attributes help us understand the structure of data.
*   In Machine Learning, knowing:`         
       *   number of samples,
       *   number of features,
       *   data type
       is critical before training a model.

```python
import numpy as np
```

# **shape – Structure of the Array**

*   shape returns a tuple representing the size of each dimension.
*   Format:        
       *   1D → (n,)
       *   2D → (rows, columns)
       *   3D → (depth, rows, columns)

```python
arr_1d = np.array([1, 2, 3, 4])
print(arr_1d.shape)
```

```python
arr_2d = np.array([[1, 2, 3],
                   [4, 5, 6]])
print(arr_2d.shape)
```

**ML meaning:**

*  (1000, 10) → 1000 samples, 10 features

# **ndim – Number of Dimensions**

*  ndim tells how many dimensions the array has.

```python
print(arr_1d.ndim)  # 1D array
print(arr_2d.ndim)  # 2D array
```

**Used to:**

*   Validate input shapes before model training
*   Handle tensors in deep learning

# **dtype – Data Type of Elements**

*   All elements in a NumPy array have the same data type.
*   Common data types:

      *   int32, int64
      *   float32, float64
      *   bool

```python
arr_int = np.array([1, 2, 3])
print(arr_int.dtype)

arr_float = np.array([1.2, 2.5, 3.8])
print(arr_float.dtype)
```

```python
# Explicitly setting dtype
arr = np.array([1, 2, 3], dtype=np.float32)
print(arr.dtype)
```

**ML relevance:**

float32 is preferred in deep learning (less memory, faster GPU computation)

# **Changing Shape and Type**

```python
arr = np.arange(12)

print(arr.shape)
print(arr.ndim)
print(arr.dtype)

reshaped = arr.reshape(3, 4)
print(reshaped.shape)
```
