# **Vectorized Operations in NumPy (No Loops)**

## **What Are Vectorized Operations?**


*   Vectorization means performing operations on entire arrays at once.
*   Avoids explicit Python loops.
*   Makes code:
    *   Faster
    *   Cleaner
    *   More readable

```python
import numpy as np
```

# **Problem with Using Loops**

*   Python loops are slow for large datasets.
*   ML models often deal with thousands or millions of values.

```python
# Using loop (not recommended)
arr = np.array([1, 2, 3, 4, 5])
result = []

for x in arr:
    result.append(x * 2)

print(result)
```

# **Vectorized Operations Example**

Same operation without loops.

```python
arr = np.array([1, 2, 3, 4, 5])
result = arr * 2
print(result)
```

*One line, faster execution.*

# **Common Vectorized Operations**

NumPy supports element-wise operations directly.

```python
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print(a + b)   # Addition
print(a - b)   # Subtraction
print(a * b)   # Multiplication
print(a / b)   # Division
```

# **Vectorized Mathematical Functions**

NumPy provides optimized math functions.

```python
arr = np.array([1, 4, 9, 16])

print(np.sqrt(arr))
print(np.log(arr))
print(np.exp(arr))
```

# **Vectorized Conditional Operations**

Apply conditions across arrays.

```python
arr = np.array([5, 15, 25, 35])

result = np.where(arr > 20, 1, 0)
print(result)
```

### **Used in:**

*   Thresholding
*   Label creation
*   Decision rules

# **ML Perspective Example**

```python
# Feature scaling using vectorization
X = np.array([10, 20, 30, 40])

X_scaled = (X - np.mean(X)) / np.std(X)
print(X_scaled)
```

*This is how preprocessing is done efficiently in ML pipelines.*
