# **Random Number Generation in NumPy (AI Relevance)**

## **Why Random Numbers Are Important in AI & ML**

Randomness plays a key role in:

*   Creating synthetic datasets
*   Weight initialization in neural networks
*   Train–test splitting
*   Data shuffling for better generalization


```python
import numpy as np
```

## **Generating Random Numbers**

NumPy provides the np.random module.

```python
# Random float between 0 and 1
rand_num = np.random.rand()
print(rand_num)
```

```python
# Random array of shape (3,)
arr = np.random.rand(3)
print(arr)
```

## **Random Integers**

Used to simulate labels or categorical values.

```python
# Random integers between 0 and 9
rand_int = np.random.randint(0, 10, size=5)
print(rand_int)
```

## **Random Numbers from Specific Distributions**
🔸 Normal (Gaussian) Distribution
*   Common in ML for noise and weight initialization.





```python
# Mean = 0, Standard Deviation = 1
normal_data = np.random.randn(5)
print(normal_data)
```

```python
# Normal distribution with custom mean and std
normal_custom = np.random.normal(loc=50, scale=10, size=5)
print(normal_custom)
```

## **Setting a Random Seed (Reproducibility)**
*   Ensures same random results every time.
*   Very important in experiments and research.

```python
np.random.seed(42)
print(np.random.rand(3))
```

## **Random Sampling from Data**

```python
data = np.array([10, 20, 30, 40, 50])

sample = np.random.choice(data, size=3, replace=False)
print(sample)
```

### **Used in:**

*   Mini-batch creation
*   Bootstrap sampling

### **ML Perspective Example**

```python
# Simulating a dataset
X = np.random.rand(100, 3)   # 100 samples, 3 features
y = np.random.randint(0, 2, size=100)  # Binary labels

print(X.shape, y.shape)
```
