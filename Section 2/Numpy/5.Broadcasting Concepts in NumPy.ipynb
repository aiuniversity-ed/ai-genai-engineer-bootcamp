# **Broadcasting Concepts in NumPy**

### **What is Broadcasting?**


*   Broadcasting allows NumPy to perform operations on arrays of different shapes.
*   Smaller arrays are automatically expanded to match the shape of larger arrays.
*   No extra memory is used for expansion.









```python
import numpy as np
```

## **Why Broadcasting is Important in ML**

*   Apply operations like:
    *   Feature scaling
    *   Bias addition
    *   Normalization
without writing loops.

## **Simple Broadcasting Example (Scalar + Array)**

```python
arr = np.array([1, 2, 3, 4])
result = arr + 10
print(result)
```

*Scalar 10 is broadcast to [10, 10, 10, 10].*




# **Broadcasting with 1D and 2D Arrays**

```python
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

vector = np.array([10, 20, 30])

result = matrix + vector
print(result)
```

*Vector is added to each row of the matrix.*

# **Broadcasting Rules (Important)**

NumPy follows two main rules:

1.   Dimensions must be equal, or
2.   One of them must be 1

Comparison starts from the rightmost dimension.

```python
# Shape: (2,3)
matrix.shape

# Shape: (3,)
vector.shape
```

*Compatible because (3,) can be stretched to (2,3).*

# **Broadcasting for Feature Scaling (ML Example)**

```python
X = np.array([[50, 200],
              [60, 180],
              [55, 220]])

mean = X.mean(axis=0)
std = X.std(axis=0)

X_scaled = (X - mean) / std
print(X_scaled)
```

*Mean and std are broadcast across all rows.*

# **Common Broadcasting Error**

```python
a = np.array([1, 2, 3])
b = np.array([1, 2])

# This will raise an error
# a + b
```

*Shapes (3,) and (2,) are not compatible.*
