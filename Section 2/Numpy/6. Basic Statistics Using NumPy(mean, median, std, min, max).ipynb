# **Basic Statistics Using NumPy(mean, median, std, min, max)**

## **Why Statistics Matter in Machine Learning**
*   Used for:
    *   Data understanding
    *   Feature scaling
    *   Outlier detection
    *   Model evaluation
*   NumPy provides fast and optimized statistical functions.

```python
import numpy as np
```

# **Mean (Average)**


*   Represents the central value of the data.
*   Formula:

	  ​mean=∑x/n

```python
data = np.array([10, 20, 30, 40, 50])
print("Mean:", np.mean(data))
```

**Used in:**

*   Normalization
*   Standardization

# **Median**


*   Middle value when data is sorted.
*   Less sensitive to outliers than mean.

```python
data = np.array([10, 20, 100, 30, 40])
print("Median:", np.median(data))
```

*Useful when data contains extreme values.*

# **Standard Deviation (std)**

*   Measures spread or variability of data.
*   High std → data widely spread
*   Low std → data clustered around mean

```python
data = np.array([10, 20, 30, 40, 50])
print("Standard Deviation:", np.std(data))
```

*Crucial for feature scaling.*

# **Minimum and Maximum**

Finds the smallest and largest values.

```python
print("Minimum:", np.min(data))
print("Maximum:", np.max(data))
```

Used for:
*   Min–Max scaling
*   Range detection

# **Statistics on 2D Arrays**

*   Use axis parameter:
    *   axis=0 → column-wise
    *   axis=1 → row-wise

```python
X = np.array([[10, 20, 30],
              [40, 50, 60]])

print("Column-wise mean:", np.mean(X, axis=0))
print("Row-wise mean:", np.mean(X, axis=1))
```

# **ML Perspective Example**

```python
# Feature-wise statistics
X = np.array([[50, 200],
              [60, 180],
              [55, 220]])

print("Mean:", X.mean(axis=0))
print("Std:", X.std(axis=0))
```

*Used before model training for preprocessing.*
