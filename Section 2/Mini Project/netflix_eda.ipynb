## Load the Dataset

## Data Cleaning
- Missing values are common in real datasets
- We fill or drop them based on importance

#Handle Missing Values
- Fill missing text-based columns

```python
# Fill missing country values
df['country'] = df['country'].fillna('Unknown')

# Fill missing director values
df['director'] = df['director'].fillna('Unknown')

# Fill missing cast values
df['cast'] = df['cast'].fillna('Unknown')

# Drop rows missing date_added since it's useful for time trends
df.dropna(subset=['date_added'], inplace=True)

# Convert date_added to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year and month added
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
```

```python
# Check missing values again
df.isnull().sum()
```

## 📊 Distribution of Movies and TV Shows on Netflix

This visualization compares the total number of Movies and TV Shows available on Netflix, helping to understand the overall content composition of the platform.


```python
plt.figure(figsize=(6,4))
sns.countplot(
    x='type',
    data=df,
    hue='type',       # explicitly assign color variable
    palette='viridis',
    legend=False      # hide legend (since x and hue are same)
)
plt.title('Number of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

## 📊 Distribution of Content Ratings on Netflix

This chart shows the frequency of different content ratings on Netflix, providing insight into the platform’s target audience and age-based content distribution.


```python
plt.figure(figsize=(10,5))
sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)
plt.title('Distribution of Ratings')
plt.xlabel('Count')
plt.ylabel('Rating')
plt.show()
```

## 📈 Netflix Content Added Over the Years

This histogram shows the number of titles added to Netflix each year, highlighting growth patterns and changes in content addition over time.


```python
plt.figure(figsize=(12,5))
sns.histplot(df['year_added'].dropna(), bins=15, kde=False, color='skyblue')
plt.title('Content Added by Year')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles Added')
plt.show()
```

## 🌍 Top 10 Countries with the Most Netflix Content

This analysis splits multi-country entries to accurately count content by country and visualizes the top 10 countries producing the highest number of titles on Netflix.


```python
# First split countries on comma and explode into one country per row
df_country = df.assign(country=df['country'].str.split(', ')).explode('country')
top_countries = df_country['country'].value_counts().nlargest(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_countries.values,
    y=top_countries.index,
    hue=top_countries.index,  # explicitly define color variable
    palette='magma',
    legend=False
)

plt.title('Top 10 Countries with Most Content')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
```

## 🎭 Top 15 Genres Available on Netflix

This visualization breaks down multi-genre entries to display the most common genres on Netflix, highlighting the platform’s dominant content categories.


```python
df_genres = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')

genre_counts = (
    df_genres['listed_in']
    .value_counts()
    .head(15)
    .reset_index()
)

genre_counts.columns = ['Genre', 'Count']
plt.figure(figsize=(10,6))
sns.barplot(
    data=genre_counts,
    x='Count',
    y='Genre'
)
plt.title('Top 15 Genres on Netflix')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

## 🎬 Distribution of Movie Durations on Netflix

This histogram illustrates the distribution of movie lengths (in minutes) available on Netflix, helping identify common runtime ranges and overall duration patterns.


```python
# Extract numeric duration safely
df['length'] = (
    df['duration']
    .str.extract(r'(\d+)')
    .astype(float)
)

# Plot movie durations
plt.figure(figsize=(10,5))
sns.histplot(
    data=df[df['type'] == 'Movie'],
    x='length',
    bins=30,
    kde=True
)
plt.title('Distribution of Movie Durations (in Minutes)')
plt.xlabel('Minutes')
plt.ylabel('Count')
plt.show()
```
