##  What is Pandas?
Pandas is a powerful Python library used for **data manipulation and analysis**.  
It provides two main data structures:
- **Series**
- **DataFrame**

These structures help us work with real-world datasets such as CSV and Excel files efficiently.

---

##  Pandas Series

###  Definition
A **Series** is a **one-dimensional labeled array** capable of holding:
- integers
- floats
- strings
- Python objects

Think of a Series as:
- A **single column** of data
- Or a **NumPy array with labels (index)**

---

###  Creating a Series


````python
import pandas as pd

# Creating a Series
marks = pd.Series([85, 90, 78, 92, 88])

marks
````

###  Key Characteristics of a Series
- One-dimensional
- Has an **index**
- Each value is associated with a label

````python
print("Values:")
print(marks.values)

print("\nIndex:")
print(marks.index)
````

###  Series with Custom Index


````python
student_marks = pd.Series([85, 90, 78], index=["Alice", "Bob", "Charlie"])

student_marks
````

## Pandas DataFrame

### Definition
A **DataFrame** is a **two-dimensional labeled data structure** with:
- rows
- columns

Think of a DataFrame as:
- An **Excel spreadsheet**
- A **table in a database**
- A collection of **multiple Series**

---

### Creating a DataFrame


````python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Marks": [85, 90, 78],
    "Passed": [True, True, True]
}

df = pd.DataFrame(data)

df
````

### Key Characteristics of a DataFrame
- Two-dimensional (rows + columns)
- Each column is a **Series**
- Can store mixed data types

````python
df.info()
````

## Series vs DataFrame (Comparison)

| Feature | Series | DataFrame |
|------|-------|---------|
| Dimension | 1D | 2D |
| Structure | Single column | Multiple columns |
| Labels | Index | Index + Columns |
| Analogy | List / Array | Table / Spreadsheet |
| Use Case | Single feature | Complete dataset |

---

## Accessing Data

###  Accessing Series Elements


````python
student_marks["Alice"]
````

### Accessing DataFrame Columns

````python
df["Marks"]
````

## When to Use What?

### Use **Series** when:
- You are working with a single feature
- You need a simple labeled array

### Use **DataFrame** when:
- You are working with real-world datasets
- You have multiple features (columns)
- You are doing data analysis or ML preprocessing

---

## Summary
- **Series** → One-dimensional data
- **DataFrame** → Two-dimensional tabular data
- Most real-world datasets are handled using **DataFrames**

# Feature Understanding: Categorical vs Numerical Data

In any dataset, features (columns) can generally be classified into **categorical** and **numerical** types. Understanding this distinction is crucial for data analysis and preprocessing.

## 1. Numerical Features
- Represent quantities or measurements.
- Can perform mathematical operations on them.
- Examples: Age, Salary, Temperature, Height.

**Sub-types:**
- **Continuous:** Can take any value within a range (e.g., height in cm).
- **Discrete:** Can only take specific values, usually integers (e.g., number of children).

## 2. Categorical Features
- Represent labels or categories.
- Cannot perform mathematical operations directly.
- Examples: Gender, Country, Product Category.

**Sub-types:**
- **Nominal:** No inherent order (e.g., Red, Blue, Green).
- **Ordinal:** Has a meaningful order (e.g., Low, Medium, High).

Understanding the type of a feature helps in:
- Choosing the right visualization (bar chart for categorical, histogram for numerical)
- Handling missing values differently
- Encoding categorical features before machine learning


````python
# Example: Identifying categorical and numerical columns in a dataset
import pandas as pd

# Sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Gender': ['F', 'M', 'M', 'M'],
    'Salary': [50000, 60000, 70000, 80000],
    'Education': ['Bachelors', 'Masters', 'PhD', 'Masters']
}

df = pd.DataFrame(data)

# Display the dataframe
print("Dataset:\n", df)

# Identify numerical columns
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
print("\nNumerical Features:", numerical_features.tolist())

# Identify categorical columns
categorical_features = df.select_dtypes(include=['object']).columns
print("Categorical Features:", categorical_features.tolist())
````

# Grouping & Aggregation: `groupby()`

When analyzing data, it's often useful to **group rows based on a feature** and then **compute summary statistics** for each group.  

Pandas provides the `groupby()` function to make this easy.

## 1. What `groupby()` does:
- Splits the data into groups based on a column or multiple columns.
- Applies an aggregation function (sum, mean, count, etc.) to each group.
- Combines the results into a new DataFrame.

## 2. Common Aggregation Functions:
- `sum()`: Sum of values in each group
- `mean()`: Average of values in each group
- `count()`: Number of rows in each group
- `max()` / `min()`: Maximum or minimum value in each group


````python
import pandas as pd

# Sample dataset
data = {
    'Department': ['HR', 'IT', 'IT', 'HR', 'Finance', 'IT', 'Finance'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace'],
    'Salary': [50000, 60000, 70000, 55000, 65000, 72000, 68000]
}

df = pd.DataFrame(data)

# Display dataset
print("Dataset:\n", df)

# Group by Department and calculate mean salary
mean_salary = df.groupby('Department')['Salary'].mean()
print("\nMean Salary by Department:\n", mean_salary)

# Group by Department and count number of employees
employee_count = df.groupby('Department')['Employee'].count()
print("\nNumber of Employees by Department:\n", employee_count)

# Group by Department and get multiple aggregations
agg_stats = df.groupby('Department')['Salary'].agg(['mean', 'max', 'min', 'sum'])
print("\nAggregated Salary Stats by Department:\n", agg_stats)
````
