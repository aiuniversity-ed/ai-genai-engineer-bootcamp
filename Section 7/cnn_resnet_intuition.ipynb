##**CNN and ResNet Intuition**

###**CNN Intuition**

A Convolutional Neural Network (CNN) is designed to process image data efficiently.

**Main Idea**

Instead of looking at the entire image at once, CNNs scan small regions of the image using filters to detect patterns.

**Simple Intuition**

Think of CNN like this:

* First layer detects edges

* Next layer detects shapes

* Deeper layers detect complex objects

**Key Components**

1. Convolution layer → extracts features

2. Pooling layer → reduces image size

3. Fully connected layer → makes final prediction




###**Hands on: CNN Model**

Import Libraries

```python
import torch
import torch.nn as nn
import torchvision.transforms as transforms
from PIL import Image
```

Load and preprocess image

```python
image = Image.open("image.png").convert("RGB")

transform = transforms.Compose([
    transforms.Resize((64,64)),
    transforms.ToTensor()
])

img = transform(image)
img = img.unsqueeze(0)  # Add batch dimension

print("Input Image Shape:", img.shape)
```

Define CNN Model

```python
class SimpleCNN(nn.Module):

    def __init__(self):
        super(SimpleCNN, self).__init__()

        self.conv1 = nn.Conv2d(3,16,3)
        self.relu = nn.ReLU()
        self.pool = nn.MaxPool2d(2,2)

        self.conv2 = nn.Conv2d(16,32,3)

        self.fc1 = nn.Linear(32*14*14,128)
        self.fc2 = nn.Linear(128,10)

    def forward(self,x):

        print("Input:", x.shape)

        x = self.conv1(x)
        print("After Conv1:", x.shape)

        x = self.relu(x)

        x = self.pool(x)
        print("After Pool1:", x.shape)

        x = self.conv2(x)
        print("After Conv2:", x.shape)

        x = self.relu(x)

        x = self.pool(x)
        print("After Pool2:", x.shape)

        x = x.view(x.size(0), -1)
        print("After Flatten:", x.shape)

        x = self.fc1(x)
        print("After FC1:", x.shape)

        x = self.fc2(x)
        print("Final Output:", x.shape)

        return x
```

Create Model


```python
model = SimpleCNN()
```

Forward Pass

```python
output = model(img)
```

Print Output

```python
print("CNN Output:", output)
```



###**ResNet Intuition**

Residual Network (ResNet) was created to solve the problem of training very deep neural networks.

**The Problem**

When networks become very deep:

* Training becomes difficult

* Gradients vanish

* Accuracy may get worse

**ResNet Idea**

* ResNet introduces skip connections (shortcut connections).

* Instead of learning a full transformation, the network learns the difference (residual) between input and output.

**Simple Intuition**

Instead of:

Input → Layer → Output

The input is added directly to the output in ResNet.

This helps the network:

* Train deeper models

* Avoid gradient problems

* Improve accuracy

###**Hands On: ResNet Model**

Import Libraries


```python
import torch
import torchvision.transforms as transforms
import torchvision.models as models
from PIL import Image
```

 Load image

```python
image = Image.open("image.png").convert("RGB")

transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor()
])

img = transform(image)
img = img.unsqueeze(0)

print("Input Shape:", img.shape)
```

Load pretrained ResNet

* We load ResNet-18 architecture which is already trained on ImageNet

* This saves training time.

```python
model = models.resnet18(pretrained=True)
```

 Forward pass
 * The image passes through multiple convolution layers and residual blocks.

* The network extracts features and produces predictions.



```python
output = model(img)
```

Print Output

```python
print("Output Shape:", output.shape)
```
