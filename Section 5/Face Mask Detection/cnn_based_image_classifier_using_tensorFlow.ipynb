# Face Mask Detection using CNN-Based Image Classifier with TensorFlow


This project aims to build a **CNN-based image classifier** using **:contentReference[oaicite:0]{index=0}** to detect whether a person is **wearing a face mask or not**.

The model is trained on a **Face Mask Detection dataset** containing two categories:

- 😷 **With Mask**
- 🙂 **Without Mask**

The system learns visual features from images and automatically classifies new images.



### What is CNN?
A **Convolutional Neural Network (CNN)** is a deep learning algorithm specially designed for **image processing**.

Unlike traditional machine learning, CNN:
- Automatically extracts important features  
- Detects edges, textures, and shapes  
- Learns patterns directly from pixel values  


### How CNN Works in This Project

### 1️⃣ Input Layer
- Images resized to **128 × 128 pixels**
- Normalized pixel values (**0 to 1**)


### 2️⃣ Convolution Layers
- Apply filters to detect:
  - Edges  
  - Facial structure  
  - Mask texture  
- Extract important visual features



### 3️⃣ MaxPooling Layer
- Reduces image size  
- Keeps important information  
- Makes the model faster and efficient  


### 4️⃣ Flatten Layer
- Converts **2D feature maps** into a **1D vector**



### 5️⃣ Fully Connected (Dense) Layer
- Learns classification patterns from extracted features



### 6️⃣ Output Layer
- Uses **Sigmoid activation**
- Outputs probability:
  - **≥ 0.5 → Without Mask**
  - **< 0.5 → With Mask**

**1. Import Libraries**

We import TensorFlow to:

- Build CNN model

- Train model

- Evaluate performance

````python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
````

**2. ImageDataGenerator**

Purpose:

- Rescales pixel values

- Splits dataset into training & validation

````python
img_size = 128
batch_size = 32

datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2
)
````

````python
train_data = datagen.flow_from_directory(
    "dataset",
    target_size=(img_size, img_size),
    batch_size=batch_size,
    class_mode='binary',
    subset='training'
)
````

````python
val_data = datagen.flow_from_directory(
    "dataset",
    target_size=(img_size, img_size),
    batch_size=batch_size,
    class_mode='binary',
    subset='validation'
)
````

**4. Build CNN Model**



## 🧱 Layers Explanation

| **Layer**      | **Purpose**                         |
|---------------|-------------------------------------|
| **Conv2D**     | Extracts important features from images (edges, textures, shapes) |
| **MaxPooling** | Reduces spatial dimensions while keeping key information |
| **Flatten**    | Converts 2D feature maps into a 1D vector |
| **Dense**      | Learns patterns and performs classification |
| **Sigmoid**    | Produces binary output (Mask / No Mask) |

- Sigmoid outputs a value between **0 and 1**  
- Used for **binary classification problems** like face mask detection

````python
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(img_size, img_size, 3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])
````

**5. Compile Model**

- Adam → Optimizer

- Binary Crossentropy → Loss function for 2 classes

- Accuracy → Performance metric

````python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
````

**6. Train Model**

- Model learns patterns from images

- Improves accuracy each epoch

````python
model.fit(
    train_data,
    epochs=10,
    validation_data=val_data
)
````

**7. Save Model**

- Saves trained model for:

- Future testing

- Real-time detection

- Deployment

````python
model.save("face_mask_model.h5")

print("Model Saved Successfully!")
````

**8. Test the model**

````python
import tensorflow as tf
import cv2
import numpy as np

model = tf.keras.models.load_model("face_mask_model.h5")

img_size = 128

img = cv2.imread("download.jpg")
img = cv2.resize(img, (img_size, img_size))
img = img / 255.0
img = np.reshape(img, (1, img_size, img_size, 3))

prediction = model.predict(img)

if prediction[0][0] > 0.5:
    print("No Mask Detected 😕")
else:
    print("Mask Detected 😷")
````
