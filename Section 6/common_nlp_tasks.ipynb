### **Step 2: Text Preprocessing**


**Text preprocessing** is the step where raw text is cleaned and transformed into a format that machine learning or NLP models can easily understand.

Raw text is often noisy and inconsistent, so preprocessing helps improve **model accuracy and efficiency**.



### 🛠️ Common Text Preprocessing Techniques

### 1️⃣ Tokenization  
- Splits text into smaller units called **tokens** (words or sentences).  
- Example:  "I love NLP" → ["I", "love", "NLP"]


```python
from nltk.tokenize import word_tokenize, sent_tokenize

text = "Natural Language Processing is amazing. It helps computers understand us!"
```

**Word Tokenization**

```python
words = word_tokenize(text)
print(f"Word Tokens: {words}")
```

**2️⃣ Sentence Tokenization**

* Sentence tokenization is the process of splitting a paragraph or text into individual sentences.

```python
sentences = sent_tokenize(text)
print(f"Sentence Tokens: {sentences}")
```

**3️⃣ Stopword Removal**
- Removes common words that add little meaning.  
- Examples: *is, the, and, a, to*  
- Example:  
"This is a book" → ["book"]

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
filtered_text = [word for word in words if word.lower() not in stop_words]

print(f"Original: {words}")
print(f"After Stopword Removal: {filtered_text}")
```

**4️⃣ Stemming / Lemmatization**  
- Reduces words to their base or root form.  
- Examples:  
- **Stemming:** `running → run`  
- **Lemmatization:** `better → good`

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer

stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

word = "rocks"
print(f"Stemming '{word}': {stemmer.stem(word)}")
print(f"Lemmatizing '{word}': {lemmatizer.lemmatize(word)}")
```

### **Step 3: Named Entity Recognition (NER)**

**Named Entity Recognition (NER)** identifies and classifies important entities in text.

**Common Entity Types**
- 👤 Person  
- 📍 Location  
- 🏢 Organization  
- 📅 Date  

We use a **pretrained Transformer-based model** to detect these entities accurately.

```python
import spacy
from spacy import displacy

text = "Apple is looking at buying a U.K. startup for $1 billion in 2026."
doc = nlp(text)

print("Entities found:")
for ent in doc.ents:
    print(f"{ent.text} -> {ent.label_}")

# To visualize it in Colab:
displacy.render(doc, style="ent", jupyter=True)
```

### **Step 4: Sentiment Analysis**

**Sentiment Analysis** determines the emotion or opinion expressed in text.

**Sentiment Categories**
- 😊 Positive  
- 😐 Neutral  
- 😞 Negative  

We use the **:contentReference[oaicite:0]{index=0} library** with a high-accuracy pretrained model.


```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

results = classifier(["I love using Google Colab for AI projects!",
                       "This error is so frustrating."])

for result in results:
    print(f"Text: {result['label']}, Score: {result['score']:.4f}")
```

### **Step 5: Part-of-Speech (POS) Tagging**


**POS Tagging** assigns a grammatical role to each word, helping understand sentence structure.

**Examples**
- Noun  
- Verb  
- Adjective  
- Adverb  

Crucial for understanding **context and meaning** in NLP tasks.


```python
text = "The quick brown fox jumps over the lazy dog."
doc = nlp(text)

print(f"{'Word':<10} | {'POS Tag':<10} | {'Explanation':<10}")
print("-" * 35)
for token in doc:
    print(f"{token.text:<10} | {token.pos_:<10} | {spacy.explain(token.pos_)}")
```

### **Step 7: Question Answering**

This task answers questions based on a provided **context**.

### How It Works
- Input: Context + Question  
- Output: Exact answer extracted from the context  

Commonly used in search engines and chatbots.

```python
# Initializing the question-answering pipeline
qa_pipeline = pipeline("question-answering")

context = "Gemini is a family of multimodal large language models developed by Google DeepMind."
question = "Who developed Gemini?"

result = qa_pipeline(question=question, context=context)

print(f"Answer: {result['answer']}")
print(f"Confidence Score: {round(result['score'], 4)}")
```

### **Step 8: Text Summarization**

**Text Summarization** shortens long text while preserving key information.

### Types of Summarization
- **Extractive:** Picks important sentences from the text  
- **Abstractive:** Generates new sentences (Transformer-based approach used)

NLP Pipeline Stages:
1. Tokenization (text → numbers)  
2. Inference (model processing)  
3. Decoding (numbers → text)

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "facebook/bart-large-cnn"

# 1. Load the tokenizer and model manually
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

text = """
The Apollo program was the third United States human spaceflight program
carried out by NASA, which succeeded in landing the first humans on the Moon.
It was first conceived during Eisenhower's administration and later dedicated
to President John F. Kennedy's goal of returning a man safely to Earth.
"""

# 2. Manual Tokenization (Convert text to tensors)
inputs = tokenizer(text, return_tensors="pt", max_length=1024, truncation=True)

# 3. Manual Inference (Generate the summary)
summary_ids = model.generate(
    inputs["input_ids"],
    max_length=50,
    min_length=20,
    length_penalty=2.0,
    num_beams=4,
    early_stopping=True
)

# 4. Manual Decoding (Convert IDs back to text)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print(f"Summary: {summary}")
```

### **Step 9: Machine Translation**


**Machine Translation** converts text from one language to another.

Modern Transformer models like **:contentReference[oaicite:2]{index=2}** support **hundreds of language pairs**.

**Example**
English → French
"Hello" → "Bonjour"

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "Helsinki-NLP/opus-mt-en-fr"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

english_text = "The quick brown fox jumps over the lazy dog."

# 1. Tokenize
inputs = tokenizer(english_text, return_tensors="pt", padding=True)

# 2. Generate Translation
translated_ids = model.generate(**inputs)

# 3. Decode
translation = tokenizer.decode(translated_ids[0], skip_special_tokens=True)

print(f"Translation: {translation}")
```
