## **Text Preprocessing**

Text Preprocessing is the process of cleaning and transforming raw text data into a structured and machine-readable format so it can be effectively used in Natural Language Processing (NLP) and machine learning models.

**Why It Is Needed**

* Raw text usually contains:

* Upper and lowercase inconsistencies

* Punctuation and special characters

* URLs and emojis

* Stopwords (e.g., is, the, and)

* Spelling variations


## **Common Text Preprocessing Steps**

Here is an example demonstrating all the text preprocessing steps from raw text to cleaned and structured output.

### **Step 1: Raw Text**


```python
text = "Hello everyone! 😊 This is a simple example showing how Text Preprocessing works in real-world NLP applications. In 2025, many companies use AI-based systems to analyze customer reviews, emails, and social media posts. Visit https://example.com for more details!!!"
print("Original Text:\n", text)
```

### **Step 2: Lowercasing**
Converting all text characters into lowercase to ensure uniformity.

```python
text_lower = text.lower()
print("\nLowercased:\n", text_lower)
```

### **Step 3: Remove URLs & Punctuation**


Eliminating punctuation marks that do not add meaningful information to the text.

```python
import re
import string

# Remove URLs
text_no_url = re.sub(r'http\S+', '', text_lower)

# Remove punctuation
text_no_punct = text_no_url.translate(str.maketrans('', '', string.punctuation))

print("\nWithout URL & Punctuation:\n", text_no_punct)
```

✔ Removes noise

✔ Keeps meaningful words

### **Step 4: Tokenization (Split into Words)**
Splitting text into smaller units such as words or sentences.

```python
tokens = text_no_punct.split()
print("\nTokens:\n", tokens)
```

✔ Breaks text into individual words

### **Step 5: Remove Stopwords**
Removing frequently occurring words that carry little semantic meaning.

```python
stopwords = {"is", "an", "how", "this"}

filtered_tokens = [word for word in tokens if word not in stopwords]
print("\nAfter Stopword Removal:\n", filtered_tokens)
```

✔ Removes common words

✔ Keeps meaningful terms

### **Step 6: Stemming**

Reducing words to their root form by trimming suffixes.

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_tokens]

print("\nAfter Stemming:\n", stemmed_words)
```

### **Step 7: Visualize word frequency**


```python
from collections import Counter

print(Counter(stemmed_words))
```
