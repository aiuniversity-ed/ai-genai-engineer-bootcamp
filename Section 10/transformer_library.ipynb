## **Transformers Library**

The Transformers library is:

A Python library built primarily on PyTorch and TensorFlow designed to make it easy to download, train, fine-tune, and use large pre-trained models

It gives developers access to thousands of models with just a few lines of code.

**Why “Transformers”?**

The name comes from the transformer architecture, introduced in:

Attention Is All You Need (2017)

Transformers use a mechanism called self-attention to understand relationships between words (or pixels, audio frames, etc.) in context.

This architecture powers many famous models like:

* BERT

* GPT-2

* RoBERTa

* T5

With the Transformers library, you can:

* Text classification


* Question answering

* Text generation

* Translation

* Summarization


**Key Components of the Transformers Library**

**Tokenizer**

Tokenizers convert text into numerical tokens that models can understand.

**Example:**

Input: "Transformers are powerful"

Tokens: [101, 19081, 2024, 3928]

**Model**

The model processes tokenized input and produces predictions.

**Examples:**

BERT for understanding text

GPT-2 for generating text.

**Pipeline**

Pipelines provide a high-level API that simplifies tasks like sentiment analysis or translation.

**Example:**

classifier = pipeline("sentiment-analysis")

The basic workflow is:

1️⃣ Input text

2️⃣ Tokenization

3️⃣ Transformer model processing

4️⃣ Output prediction

**Import Libraries**


* pipeline → simple way to use pretrained models

* AutoTokenizer → converts text to tokens

* AutoModel → loads pretrained transformer models.


```python
from transformers import pipeline, AutoTokenizer, AutoModel
```

**Tokenization Demo**

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "Transformers are powerful models for natural language processing."

tokens = tokenizer(text)

print(tokens)
```

* Transformers cannot read text directly

* Text is converted to token IDs.

* This uses BERT.

**Sentiment Analysis**

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")

text = "I really enjoy learning about transformer models."

result = classifier(text)

print(result)
```

**Text Generation**


* Model predicts next words in a sequence.

* Uses GPT-2.

```python
generator = pipeline("text-generation", model="gpt2")

prompt = "Artificial intelligence will"

output = generator(prompt, max_length=30)

print(output)
```

**Question Answering**

* Model reads a paragraph and finds the answer from context.

```python
qa_pipeline = pipeline("question-answering")

context = """
Transformers are deep learning models introduced by Google researchers in 2017.
They use self-attention mechanisms to process sequential data.
"""

question = "When were transformers introduced?"

result = qa_pipeline(question=question, context=context)

print(result)
```

**Text Summarization**

```python
from transformers import pipeline

summarizer = pipeline("text-generation", model="facebook/bart-large-cnn")

article = """
Transformers have revolutionized natural language processing.
They allow models to understand relationships between words using attention mechanisms.
This architecture is now used in many applications including chatbots, translation systems, and text summarization.
"""

summary = summarizer(article, max_length=40)

print(summary)
```

**Machine Translation**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "Helsinki-NLP/opus-mt-en-fr"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

text = "Machine learning is transforming technology."

inputs = tokenizer(text, return_tensors="pt")

outputs = model.generate(**inputs)

translation = tokenizer.decode(outputs[0], skip_special_tokens=True)

print("Translated text:", translation)
```
