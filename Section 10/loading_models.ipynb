# **🤗 Hugging Face – Loading Models**

**Hugging Face** is a popular open-source ecosystem that makes it easy to use state-of-the-art **pretrained machine learning models** for tasks such as text classification, translation, summarization, question answering, and more.

Using the **Transformers** library, we can load powerful models that are already trained on massive datasets—**without training from scratch**.

---

### Why Use Hugging Face Models?

- **Pretrained & ready to use**  
- Supports **Transformer-based architectures** (BERT, GPT, T5, etc.)
- Simple and consistent API
- Works seamlessly with **PyTorch** and **TensorFlow**
- Large community and thousands of shared models

---

### 🔹 What Does “Loading a Model” Mean?

Loading a model involves:
1. Downloading the pretrained weights from the **Hugging Face Model Hub**
2. Loading the corresponding **tokenizer**
3. Preparing the model for inference or fine-tuning

Once loaded, the model can directly process text inputs and generate predictions.

---

### 🔹 Typical Components Loaded

- **Tokenizer** → Converts raw text into numerical tokens
- **Model** → Neural network with pretrained parameters
- **Configuration** → Model architecture and settings

---

### Example Tasks Using Loaded Models

- Sentiment Analysis
- Named Entity Recognition (NER)
- Text Summarization
- Machine Translation
- Question Answering

---



> Hugging Face allows us to focus on **solving problems**, not on **building models from scratch**.

## **i. AutoTokenizer**

👉 Converts **text → tokens → numerical IDs**

- Transformers don’t understand text.

- They understand numbers.

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = "Hugging Face is amazing!"
tokens = tokenizer(text)

print(tokens)
```

| **Field** | **Meaning** |
|----------|-------------|
| **input_ids** | Numerical IDs representing tokens |
| **attention_mask** | Indicates which tokens are real (1) vs padding (0) |
| **token_type_ids** | Separates sentences (used mainly in BERT-like models) |

> These fields are passed directly to transformer models as inputs.

## **ii. AutoModel**

👉 **Loads only the base transformer model (no task-specific head)**

Think of this as:

> **Just the brain — no classifier attached**

Used when:
- You want embeddings
- You plan to add your own custom head
- You are doing feature extraction

```python
from transformers import AutoModel, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")

text = "Hello, how are you?"

tokens = tokenizer(text, return_tensors="pt")

outputs = model(**tokens)

print(outputs.last_hidden_state.shape)
```

| **Dimension** | **Meaning** |
|--------------|-------------|
| 1 | Batch size |
| 7 | Number of tokens |
| 768 | Hidden embedding size |

## **iii. AutoModelForTask**

👉 **Loads the base model + a task-specific head**

This is used for direct prediction tasks.

### 🔍 Common Variants

| **Class** | **Use Case** |
|---------|--------------|
| **AutoModelForSequenceClassification** | Sentiment analysis, spam detection |
| **AutoModelForTokenClassification** | Named Entity Recognition (NER) |
| **AutoModelForCausalLM** | GPT-style text generation |
| **AutoModelForQuestionAnswering** | Question Answering systems |

> These models are ready for training or inference on specific tasks.


```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

model_name = "distilbert-base-uncased-finetuned-sst-2-english"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

text = "I love Hugging Face!"

inputs = tokenizer(text, return_tensors="pt")
outputs = model(**inputs)

logits = outputs.logits
predicted_class = torch.argmax(logits).item()

print(predicted_class)
```

## 🎯 Loading Models for Specific Tasks

In Hugging Face, **different tasks require different model heads**.  
Instead of manually adding layers, Hugging Face provides **task-specific `AutoModel` classes** that automatically attach the correct head on top of the base transformer.

---

###  Why Multiple `AutoModel` Classes?

- Different tasks → different output formats
- Each class adds a **task-specific head**
- Simplifies training and inference
- Reduces implementation errors

---

### 📚 Task → AutoModel Mapping

| **Task** | **AutoModel Class** |
|--------|---------------------|
| **Text Classification** | `AutoModelForSequenceClassification` |
| **Token Classification** | `AutoModelForTokenClassification` |
| **Question Answering** | `AutoModelForQuestionAnswering` |
| **Text Generation** | `AutoModelForCausalLM` |
| **Seq2Seq (T5 / BART)** | `AutoModelForSeq2SeqLM` |





## **Text Classification**

`AutoModelForSequenceClassification`

Used for:

- Sentiment analysis

- Spam detection

- Topic classification

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

model_name = "distilbert-base-uncased-finetuned-sst-2-english"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

text = "This course is amazing!"
inputs = tokenizer(text, return_tensors="pt")

with torch.no_grad():
    outputs = model(**inputs)

probs = torch.softmax(outputs.logits, dim=1)

predicted_class_id = torch.argmax(probs, dim=1).item()
predicted_label = model.config.id2label[predicted_class_id]
confidence = probs[0][predicted_class_id].item()

print(f"Predicted Sentiment: {predicted_label}")
print(f"Confidence: {confidence:.4f}")
```

## **Token Classification**

`AutoModelForTokenClassification`

Used for:

- Named Entity Recognition (NER)

- POS tagging

- Chunking

```python
from transformers import pipeline

ner = pipeline(
    "ner",
    aggregation_strategy="simple"
)

text = "Elon Musk founded SpaceX in California."

results = ner(text)

for entity in results:
    print(f"Entity: {entity['word']}")
    print(f"Label: {entity['entity_group']}")
    print(f"Confidence: {entity['score']:.4f}")
    print("-" * 30)
```

## **Question Answering**

`AutoModelForQuestionAnswering`

Used for:

- Extractive QA

- Context-based answer extraction

```python
from transformers import pipeline

qa = pipeline("question-answering")

context = "Hugging Face is a company based in New York."
question = "Where is Hugging Face based?"

result = qa(question=question, context=context)

print(result)
```

## **Text Generation**

`AutoModelForCausalLM`

Used for:

- GPT-style generation

- Chat models

- Code generation

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "gpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

inputs = tokenizer("Artificial intelligence is", return_tensors="pt")

outputs = model.generate(**inputs, max_length=20)

print(tokenizer.decode(outputs[0]))
```

## **Seq2Seq Models (T5, BART)**

`AutoModelForSeq2SeqLM`

Used for:

- Translation

- Summarization

- Paraphrasing

- Instruction-following

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

model_name = "t5-small"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

text = "summarize: Hugging Face develops tools for machine learning and AI research. The company provides widely used open-source libraries that make it easier to build, train, and deploy natural language processing models. Its Transformers library allows developers to access state-of-the-art pretrained models with just a few lines of code."

inputs = tokenizer(text, return_tensors="pt")

outputs = model.generate(**inputs, max_length=30)

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```

## **4. Loading Models to GPU / Device Management**

Why Use This?

* Automatically distributes model across GPUs

* Uses mixed precision

* Required for large LLMs

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model.to(device)

inputs = tokenizer("Hello world", return_tensors="pt").to(device)
```

## **5. Loading in 8-bit / 4-bit (Memory Optimization)**

- 🚀 **Up to 4× memory reduction**
- 🧠 Enables loading **large models on smaller GPUs**
- 💰 Reduces hardware requirements

Commonly used when:
- Working with LLMs
- Running inference on limited GPU memory
- Deploying cost-efficient services

```python
from transformers import BitsAndBytesConfig

bnb_config = BitsAndBytesConfig(load_in_8bit=True)

model = AutoModelForCausalLM.from_pretrained(
    "gpt2",
    quantization_config=bnb_config,
    device_map="auto"
)
```

## **6. Loading From Local Directory**



Loading models from a **local directory** is common in real-world and production scenarios.

| **Scenario** | **Why** |
|-------------|--------|
| Fine-tuned model | Reload the model after training |
| Air-gapped servers | No internet access |
| Production deployment | Use a stable, versioned model artifact |
| Docker containers | Copy the model into the container image |



Loading works if the directory includes:

- `config.json`
- `pytorch_model.bin` (or `model.safetensors`)
- Tokenizer files (`tokenizer.json`, `vocab.txt`, etc.)

> If these files exist, Hugging Face treats the folder just like a model from the Hub.

```python
model = AutoModelForSequenceClassification.from_pretrained("./saved_model")
tokenizer = AutoTokenizer.from_pretrained("./saved_model")
```
