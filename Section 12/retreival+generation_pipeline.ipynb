## **Retrieval + Generation Flow**

RAG consists of two primary stages that occur in real-time when a user asks a question:

**Retrieval:** The system searches a specialized knowledge base (like a vector database containing your PDFs or company manuals) to find snippets of text most relevant to the user's query.


**Generation:** These relevant snippets are added to the original user prompt as "context." The LLM then uses this context to generate a final, grounded answer.

To implement a simple Retrieval-Augmented Generation (RAG) flow, you need a collection of text documents (the knowledge base) and a basic script to link them.

### **1. Simple Dataset**

This dataset represents a company FAQ (e.g., HR policy) in a structured dictionary format, which is easily converted to a Pandas DataFrame or used directly.


```python
dataset = [
    {"id": 1, "topic": "Vacation", "context": "Employees get 20 days of paid vacation per year."},
    {"id": 2, "topic": "Remote Work", "context": "Remote work is allowed up to 3 days a week, with Tuesdays and Thursdays mandatory in-office."},
    {"id": 3, "topic": "Insurance", "context": "Health insurance covers dental and vision after 90 days of employment."},
    {"id": 4, "topic": "Office Hours", "context": "Core office hours are 10:00 AM to 4:00 PM."},
    {"id": 5, "topic": "Laptop", "context": "A MacBook Pro or Dell XPS is provided upon joining."}
]
```

### **2. Retrieval + Generation Flow Code**

This code uses semantic similarity to find the right chunk and a placeholder for an LLM (like OpenAI or LangChain) to generate an answer.


### Step 1: Import Libraries

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

### Step 2: Create a Simple Dataset (Knowledge Base)

Knowledge Base is a simple list of documents or FAQs.

```python
knowledge_base = [d['context'] for d in dataset]
```

### Step 3: Vectorisation

Vectorization is used for Converting text to embeddings (here, simplified using TF-IDF).

```python
vectorizer = TfidfVectorizer()
doc_embeddings = vectorizer.fit_transform(knowledge_base)
```

### User Query

```python
query = "How many days can I work from home?"
```

### Step 4: Retrieval

Retriever helps in calculating cosine similarity to find the best match.

```python
query_embedding = vectorizer.transform([query])
similarities = cosine_similarity(query_embedding, doc_embeddings)
best_match_idx = np.argmax(similarities)
relevant_context = knowledge_base[best_match_idx]
```

### Step 5: Generation (Simulated prompt to LLM)
In a real scenario, you would call OpenAI/LangChain here:

    # response = llm.invoke(prompt)
    
Here we use Simple simulation.

```python
def generate_answer(query, context):
    prompt = f"""
    Use the following context to answer the question.
    Context: {context}
    Question: {query}
    Answer:
    """

    answer = f"Based on the policy: {context}"
    return answer
```

### Step 6: Output Results

```python
print(f"Query: {query}")
print(f"Retrieved Context: {relevant_context}")
print(f"Generated Answer: {generate_answer(query, relevant_context)}")
```

Key Benefits of RAG:

* **Reduces "Hallucinations":** By grounding answers in specific facts, the model is less likely to make things up.
* **Up-to-Date Information:** You can update the knowledge base with the latest news or internal data without the massive cost of retraining the entire LLM.
* **Source Attribution:** RAG allows the model to cite exactly which document it used, increasing user trust.
* **Privacy & Security:** Sensitive data can be kept in a local database and only shared with the model at query time, rather than being "baked into" the model's public training.
